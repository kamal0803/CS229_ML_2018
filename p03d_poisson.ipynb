```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('ds4_train.csv')
```

```python
df_train
```

```python
X = df_train.iloc[:, :-1].values
y = df_train.iloc[:, -1:].values
m, n = X.shape
```

```python
# taking random values of theta, and transforming it to a n x 1 matrix
theta = np.ones(n).reshape(-1, 1)
```

```python
theta.shape
```

```python
def compute_natural_parameter(X, theta):

  eta = np.exp(np.dot(X, theta))

  return eta
```

```python
alpha = 1e-7
```

```python
for i in range(100000):

  # gradient ascent algorithm to iteratively calculate theta
  theta = theta + alpha*np.dot(X.T, (y - compute_natural_parameter(X, theta)))/m
```
