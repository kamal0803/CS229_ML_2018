```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ds3_train.csv')
```

```python
def add_intercept(x):
  new_x = np.zeros((x.shape[0], x.shape[1] + 1), dtype=x.dtype)
  new_x[:, 0] = 1
  new_x[:, 1:] = x

  return new_x
```

```python
def sigmoid_func(X, theta):

  return 1/(1 + np.exp(-(np.dot(X, theta))))
```

```python
alpha = 0.001
```

## Coding Problem - c

```python
X = df.iloc[:, 1:3].values
y = df.iloc[:, 0:1].values
```

```python
X = add_intercept(X)
```

```python
theta = np.array([1, 1, 1]).reshape(3, 1)
```

```python
for i in range(100000):
  theta = theta + np.dot(X.T, (y - sigmoid_func(X, theta)))*alpha
```

```python
theta
```

```python
theta0 = theta[0]
theta1, theta2 = theta[1], theta[2]

# Generate data points and plot them
plt.scatter(X[:, 1], X[:, 2], c=y[:, 0], cmap=plt.cm.RdBu, edgecolors='k')

# Define the decision boundary equation
x1_vals = np.linspace(X[:, 1].min(), X[:, 1].max(), 100)
decision_boundary = - (theta0 + theta1 * x1_vals) / theta2

# Plot the decision boundary
plt.plot(x1_vals, decision_boundary, color='black', linestyle='--')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Logistic Regression Decision Boundary')

plt.show()
```
