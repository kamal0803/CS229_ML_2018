```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ds2_train.csv')
```

```python
X = df.iloc[:, :-1].values
```

```python
y = df.iloc[:, -1:].values
```

```python
m = len(df.index)
```

```python
y1 = len(df[df['y'] == 1.0].index)
phi = y1/m
```

```python
mu_0 = [df[df['y'] == 0.0]['x_1'].sum() / (m - y1), df[df['y'] == 0.0]['x_2'].sum() / (m - y1)]
```

```python
mu_1 = [df[df['y'] == 1.0]['x_1'].sum() / y1, df[df['y'] == 1.0]['x_2'].sum() / y1]
```

```python
mu_0
```

```python
mu_1
```

```python
df
```

```python
for idx, row in df.iterrows():
  if row['y'] == 0:
    df.at[idx, 'x_1_adjusted'] = row['x_1'] - mu_0[0]
    df.at[idx, 'x_2_adjusted'] = row['x_2'] - mu_0[1]
  else:
    df.at[idx, 'x_1_adjusted'] = row['x_1'] - mu_1[0]
    df.at[idx, 'x_2_adjusted'] = row['x_2'] - mu_1[1]
```

```python
df
```

```python
y_0_array = df[df['y'] == 0][['x_1_adjusted', 'x_2_adjusted']].to_numpy()
y_1_array = df[df['y'] == 1.0][['x_1_adjusted', 'x_2_adjusted']].to_numpy()
```

```python
sigma = y_0_array.T.dot(y_0_array) + y_1_array.T.dot(y_1_array)
```

```python
sigma
```

```python
sigma_inv = np.linalg.inv(sigma)
```

```python
theta = np.zeros(3)
```

```python
mu_0 = np.array(mu_0)
mu_1 = np.array(mu_1)
```

```python
theta[0] = 0.5 * (mu_0 + mu_1).dot(sigma_inv).dot(mu_0 - mu_1) - np.log((1 - phi) / phi)
theta[1:] = sigma_inv.dot(mu_1 - mu_0)
```
