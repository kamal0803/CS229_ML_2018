```python
import pandas as pd
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('ds5_train.csv')
```

```python
df_valid = pd.read_csv('ds5_train.csv')
```

```python
# Intercept term added to data set

df_train.insert(0, 'Intercept',1)
df_valid.insert(0, 'Intercept', 1)
```

```python
X_train = df_train.iloc[:, :-1].values
y_train = df_train.iloc[:, -1:].values

X_valid = df_valid.iloc[:, :-1].values
y_valid = df_valid.iloc[:, -1:].values
```

```python
def model_prediction(X, y, tau):

  m, n = X.shape
  w = []
  y_pred = np.zeros(m)
  for i in range(m):
    w.append(np.diag(np.exp(-np.sum((X[i] - X)**2, axis=1)/(2*tau*tau))))
    theta = np.dot(np.dot(np.dot(np.linalg.inv(np.dot(np.dot(X.T, w[i]) , X)), X.T), w[i]), y)
    # since Locally Weighted Regression (LWR) is a non-parametric regression algorithm, it wont have fixed set of theta, it would grow
    # linearly with data.

    y_pred[i] = np.dot(X[i], theta)

  return y_pred
```

```python
def calc_mse(y_pred, y):

  mse = np.mean((y_pred.reshape(-1, 1) - y)**2)

  return mse
```

```python
def plot_predictions(X_train, y_train, X_valid, y_pred):
  plt.figure()
  plt.plot(X_train, y_train, 'bx', linewidth=2)
  plt.plot(X_valid, y_pred.reshape(-1,1), 'ro', linewidth=2)
```

```python
tau = 0.5
```

```python
y_pred = model_prediction(X_valid, y_valid, tau)
```

```python
calc_mse(y_pred, y_valid)
```

```python
plot_predictions(X_train, y_train, X_valid, y_pred)
```
